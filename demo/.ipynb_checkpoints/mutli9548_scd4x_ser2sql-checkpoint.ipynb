```python
import serial
import time

# 配置串口
ser = serial.Serial('COM6', 115200, timeout=1)  # 请根据实际情况修改COM端口
time.sleep(2)  # 等待串口初始化

# 打开文件准备写入
with open('sensor_data.txt', 'w') as f:
    while True:
        # 读取串口数据
        line = ser.readline().decode('utf-8').rstrip()
        if line:
            print(line)  # 打印到控制台
            f.write(line + '\n')  # 写入文件
            f.flush()  # 确保数据写入文件

# 关闭串口
ser.close()
```

```python
import serial
import mysql.connector
from datetime import datetime

# 配置串口
ser = serial.Serial('COM6', 115200, timeout=1)  # 请根据实际情况修改COM端口

# 配置数据库连接
db_config = {
    'user': 'fudan',  # 替换为你的数据库用户名
    'password': '',  # 替换为你的数据库密码
    'host': 'localhost',
    'database': 'fudan'
}

# 连接到数据库
db_connection = mysql.connector.connect(**db_config)
cursor = db_connection.cursor()

# 打开串口
ser.flushInput()

try:
    while True:
        # 读取串口数据
        line = ser.readline().decode('ascii').rstrip()
        if line:

            # 分割数据并转换为相应的类型
            sensor_id, tmp, hmd, co2 = map(float, line.split(','))
            id = int(sensor_id)  # 转换为整数
            
            # 构造插入SQL语句
            insert_query = (
                "INSERT INTO scd4x_scd30 (id, temperature, humidity, CO2) "
                "VALUES (%s, %s, %s, %s)"
            )
            data = (id, tmp, hmd, co2)
            cursor.execute(insert_query, data)
            
            # 提交到数据库执行
            db_connection.commit()
            
            print(f"Inserted data: {data}")
    pass
finally:
    # 关闭串口
    ser.close()
    # 关闭数据库连接
    cursor.close()
    db_connection.close()
```
